```python
# Lets start by getting our data:

import csv
import os
import cv2

images = []
images_center = []
images_left = []
images_right = []

steering_data = []
steering_data_center = []
steering_data_left = []
steering_data_right = []

correction = 0.2

# Get the data lines for each data file
for d in os.listdir("data/"):
    if d != "driving_log.csv" and d != "IMG":
        with open("data/" + d + "/driving_log.csv") as csvFile:
            reader = csv.reader(csvFile)
            
            for line in reader:
                # Center image and data
                filename = line[0].split("\\")[-1]
                image = cv2.imread("data/" + d + "/IMG/" + filename)
                images_center.append(image)
                steering = float(line[3])
                steering_data_center.append(steering)
                
                #left
                filename = line[1].split("\\")[-1]
                image = cv2.imread("data/" + d + "/IMG/" + filename)
                images_left.append(image)
                steering = float(line[3]) + correction
                steering_data_left.append(steering)
            
                # Right
                filename = line[2].split("\\")[-1]
                image = cv2.imread("data/" + d + "/IMG/" + filename)
                images_right.append(image)
                steering = float(line[3]) - correction
                steering_data_right.append(steering)
                
images.extend(images_center)
images.extend(images_right)
images.extend(images_left)

steering_data.extend(steering_data_center)
steering_data.extend(steering_data_right)
steering_data.extend(steering_data_left)
```

```python
#Visualize the data

%matplotlib inline
import matplotlib.pyplot as plt

plt.imshow(images[100])
print("Steering angle: {:.2f}".format(steering_data[100]))
```

```python
# Define our training data and labels

import numpy as np

x_train = np.array(images)
y_train = np.array(steering_data)

print("{} training images".format(len(x_train)))
print("{} training labels".format(len(y_train)))
```

```python
# Data augmentation
# For each of our images, let's flip the image left to right.
# This provides us with double the training data, but also
# will remove any bias towards the left or right

x_train_aug, y_train_aug = [], []

for image, label in zip(x_train, y_train):
    x_train_aug.append(image)
    y_train_aug.append(label)
    
    x_train_aug.append(cv2.flip(image, 1))
    y_train_aug.append(label * -1.0)
    
print("{} augmented training images".format(len(x_train_aug)))
print("{} augmented training labels".format(len(y_train_aug)))
```

```python
# Build the model

from keras.layers import Conv2D, Lambda, Cropping2D, Flatten, Dense
from keras.models import Sequential

def build_model(input_shape, output):
    model = Sequential()
    model.add(Lambda(lambda x: x / 255 - 0.5, input_shape=input_shape))
    model.add(Cropping2D(cropping=((70, 25), (0, 0))))  # Crops off the hood and sky
    model.add(Conv2D(24, (5, 5), strides=(2, 2), activation="relu"))
    model.add(Conv2D(36, (5, 5), strides=(2, 2), activation="relu"))
    model.add(Conv2D(48, (5, 5), strides=(2, 2), activation="relu"))
    model.add(Conv2D(64, (3, 3), activation="relu"))
    model.add(Conv2D(64, (3, 3), activation="relu"))
    model.add(Flatten())
    model.add(Dense(100))
    model.add(Dense(50))
    model.add(Dense(10))
    model.add(Dense(1))
    
    model.compile(loss="mse", optimizer="adam")
    return model
    
```

```python
input_shape = x_train_aug[0].shape
print("Input shape: {}".format(input_shape))

model = build_model(input_shape, 1)
```

```python
# Training
x_train_aug = np.array(x_train_aug)
y_train_aug = np.array(y_train_aug)

x_split = np.split(x_train_aug, 4)
y_split = np.split(y_train_aug, 4)

for x,y in zip(x_split, y_split):
    model.fit(x, y, validation_split=0.2, shuffle=True, verbose=1, batch_size = 256)
    model.save("model.h5")
```
